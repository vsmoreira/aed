```python
import pandas as pd
```

```python
resultados_n_450 = pd.read_csv('../resources/resultados_n_450.csv')
resultados_n_450_vies = pd.read_csv('../resources/resultados_n_450_ponderado.csv')
```

## Sem atenuador dos custos

```python
# Estatísticas a respeito do comportamento do fator tempo_servico antes da realocação
resultados_n_450[['classificacao_anterior','tempo_servico']].groupby('classificacao_anterior').describe()
```

```python
# Médias dos tempos antes da realocação
resultados_n_450[['classificacao_anterior', 'tempo_servico','tempo_na_lotacao','tempo_fora','tempo_sede']].groupby('classificacao_anterior').mean()
```

```python
# Estatísticas a respeito do comportamento do fator tempo_servico após realocação
resultados_n_450[['classificacao_alocada','tempo_servico']].groupby('classificacao_alocada').describe()
```

```python
# Médias dos tempos após realocação
resultados_n_450[['classificacao_alocada', 'tempo_servico','tempo_na_lotacao','tempo_fora','tempo_sede']].groupby('classificacao_alocada').mean()
```

## Utilização de atenuador nos custos

```python
# Estatísticas a respeito do comportamento do fator tempo_servico antes da realocação
resultados_n_450_vies[['classificacao_anterior','tempo_servico']].groupby('classificacao_anterior').describe()
```

```python
# Estatísticas a respeito do comportamento do fator tempo_servico após realocação
resultados_n_450_vies[['classificacao_alocada','tempo_servico']].groupby('classificacao_alocada').describe()
```
